```python
#creating function which will take the information about student
def info_detail():
    info = {}
    #taking inputs for dictionary
    student_id = input("student id:")
    name = input("name:").title()
    father_name = input("Father name:").title()
    Class = input("Enter class:").upper()
    Session = input("Which session:")
    
    #adding input values to dictionary
    info["Student id"] = student_id
    info["Name"] = name
    info["Father name"] = father_name
    info["Class"] = Class
    info["Session"] = Session
    print("\n\t\t\t***MARK SHEET***\n STUDENT INFORMATION")
    
    #printing the keys and values from dictionary where student information is stored
    for key,value in info.items():
           print("{:11}{:>13}".format(key,value))

#Subject names and marks
subjects = ["English","Urdu","Mathematics","Physics","Chemistry","Pak_studies","Islamiat"]
obt_marks = []
total_marks = [100,100,100,100,100,50,50]

#to take input for the obtain marks 
for sub in subjects:
    if sub == "Pak_studies":
        marks = int(input("Enter obtained marks out of 50 for {:15}".format(sub)))
        obt_marks.append(marks)
    elif sub == "Islamiat":
        marks = int(input("Enter obtained marks out of 50 for {:15}".format(sub)))
        obt_marks.append(marks)
    else:    
        marks = int(input("Enter obtained marks for {:15}".format(sub)))
        obt_marks.append(marks)

info_detail()
print("\nSUBJECTS\t\tTOTAL MARKS\tOBTAINED MARKS")
for s,o,t in zip(subjects,obt_marks,total_marks):
    print("{:11}{:24}{:19}".format(s,t,o))

def sub_wise_checking():
    f_sub = []
    s_per=[]
    a = 0
    while a<7:
        o = obt_marks[a]
        t = total_marks[a]
        p = (o/t)*100
        s_per.append(p)
        if p<33:
            f_sub.append(a)
        a = a+1
    
    if any(i < 33 for i in s_per) == True:
        g = "Fail"
        remarks = "Failure is the key to success,It shows that at least you tried"
        grand_t_total=sum(total_marks)
        grand_o_total = sum(obt_marks)
        
        
    elif all(i > 33 for i in s_per) == True:
        print("All papers are clear")
        grand_t_total=sum(total_marks)
        grand_o_total = sum(obt_marks)
        per = grand_o_total/grand_t_total*100

        #for grading policy        


        if per >= 80:
            g= "A+"
            remarks = "Suplended Effort"
        elif per >= 70:
            g = "A"
            remarks = "Well done, Keep it up!"
        elif per >= 60:
            g = "B"
            remarks = "Work hard,you can do better."
        elif per >= 50:
            g= "C"
            remarks = "Not Stisfactory.But you are in learning period, so learn it now rather than it become so late"
        elif per >= 33:
            g = "Pass"
            remarks = "Disappointing,In sha ALLAH it is the last time you will surprize me next time with great improvement"

        else:
            g = "Fail"
            remarks = "Failure is the key to success,It shows that at least you tried"

        print("\nOverall total marks={}\nTotal Obtained Marks={}\nPercentage={:.4}%\nGrade={}\nRemarks:{}".format(grand_t_total,grand_o_total,per,g,remarks))
    failure = []
    for i in f_sub:
        fail_sub_name=subjects[i]
        failure.append(fail_sub_name)
    if len(failure) == 1:
        print(f"\n{failure} paper is failed")
        print("\nOverall total marks={}\nTotal Obtained Marks={}\nGrade={}\nRemarks:{}".format(grand_t_total,grand_o_total,g,remarks))
    elif len(failure) > 1:
        print(f"\n{failure} papers are failed")
        print("\nOverall total marks={}\nTotal Obtained Marks={}\nGrade={}\nRemarks:{}".format(grand_t_total,grand_o_total,g,remarks))

sub_wise_checking()
    
```
