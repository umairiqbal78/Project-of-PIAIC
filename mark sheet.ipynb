```python
#creating function which will take the information about student
def info_detail():
    info = {}
    #taking inputs for dictionary
    student_id = input("student id:")
    name = input("name:").title()
    father_name = input("Father name:").title()
    Class = input("Enter class:").title()
    Session = input("Which session:")
    
    #adding input values to dictionary
    info["Student id"] = student_id
    info["Name"] = name
    info["Father name"] = father_name
    info["Class"] = Class
    info["Session"] = Session
    print("\n\t\t\t***MARK SHEET***\n STUDENT INFORMATION")
    
    #printing the keys and values from dictionary
    for key,value in info.items():
           print("{:11}{:>13}".format(key,value))


subjects = ["English","Urdu","Mathematics","Physics","Chemistry","Pak_studies","Islamiat"]
obt_marks = []
total_marks = [100,100,100,100,100,50,50]

for sub in subjects:
        marks = int(input("Enter obtained marks for {:15}".format(sub)))
        obt_marks.append(marks)

info_detail()
print("\nSUBJECTS\t\tTOTAL MARKS\tOBTAINED MARKS")
for s,o,t in zip(subjects,obt_marks,total_marks):
    print("{:11}{:24}{:19}".format(s,t,o))

    
grand_t_total=sum(total_marks)
grand_o_total = sum(obt_marks)
per = grand_o_total/grand_t_total*100

#for grading policy        


if per >= 80:
    g= "A+"
    print(g) 
elif per >= 70:
    g = "A"
    
elif per >= 60:
    g = "B"
elif per >= 50:
    g= "C"
elif per >= 33:
    g = "Pass"

else:
    g = "Fail"



print("\nOverall total marks={}\nTotal Obtained Marks={}\nGrade={}".format(grand_t_total,grand_o_total,g))
```
